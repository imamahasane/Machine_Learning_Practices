```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
x_mean
```

```python
y_mean
```

```python
rise = np.sum((x - x_mean) * (y - y_mean))
```

```python
rise
```

```python
run = np.sum((x - x_mean) ** 2)
```

```python
run
```

```python
slop = rise / run
```

```python
slop
```

```python
intercept = y_mean - (slop * x_mean)
```

```python
intercept
```

```python
plt.scatter(x, y, color = 'red', label = 'Data Observation')
plt.plot(x, slop * x + intercept, color = 'blue', label = 'SLR')
plt.legend()
plt.xlabel('Independent Data Point')
plt.ylabel('Dependent Data Point')
plt.title('SLR')
plt.show()
```
